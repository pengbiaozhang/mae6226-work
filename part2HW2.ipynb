```python
import math
import numpy
from matplotlib import pyplot
# embed the figures into the notebook
%matplotlib inline
```

```python
N=100
c=1
R=1.5
arfa=numpy.linspace(0, 2*numpy.pi, N)
z=R*numpy.sin(arfa)+R*numpy.cos(arfa)*1j
zeta=z+c**2/z
```

```python
size = 10
x_start, x_end = -2.0, 2.0            
y_start, y_end = -2.0, 2.0 
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-3, 3)
pyplot.ylim(-2, 2)
pyplot.plot(zeta.real,zeta.imag)
pyplot.grid(1)
```

```python
N=100
c=1
R=1.5
arfa=numpy.linspace(0, 2*numpy.pi, N)
b=R*numpy.sin(arfa)+R*numpy.cos(arfa)*1j
z=b-0.2
zeta=z+c**2/z
```

```python
size = 10
x_start, x_end = -2.0, 2.0            
y_start, y_end = -2.0, 2.0 
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-3, 3)
pyplot.ylim(-2, 2)
pyplot.plot(zeta.real,zeta.imag)
pyplot.grid(1)
```

```python
N=100
c=1
x_circle=-0.1
y_circle=0.1
circle=x_circle+y_circle*1j
R=math.sqrt((c-x_circle)**2+y_circle**2)
arfa=numpy.linspace(0, 2*numpy.pi, N)
b=R*numpy.sin(arfa)+R*numpy.cos(arfa)*1j
z=b+circle
zeta=z+c**2/z
```

```python
size = 10
x_start, x_end = -2.0, 2.0            
y_start, y_end = -2.0, 2.0 
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-3, 3)
pyplot.ylim(-2, 2)
pyplot.plot(zeta.real,zeta.imag)
pyplot.grid(1)
```
